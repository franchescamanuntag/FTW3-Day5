```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas_profiling as pp
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split # split dataset
from sklearn.metrics import mean_squared_error as mse # Measurement metric
```

```python
data = pd.read_csv("nyc-rolling-sales.csv")
```

```python
data['SALE_PRICE'].head()
```

```python
data['SALE_PRICE'] = pd.to_numeric(data['SALE_PRICE'], errors='coerce')
```

```python
data['SALE_PRICE'].head()
```

```python
data['GROSS_SQUARE_FEET'] = pd.to_numeric(data['GROSS_SQUARE_FEET'], errors='coerce')
```

```python
data['GROSS_SQUARE_FEET'].head()
```

```python
data['LAND_SQUARE_FEET'] = pd.to_numeric(data['LAND_SQUARE_FEET'], errors='coerce')
```

```python
data['LAND_SQUARE_FEET'].head()
```

```python
data['SALE_PRICE'].isna().sum()
```

```python
data[('SALE_PRICE')] = data['SALE_PRICE'].fillna(0)
```

```python
data[('GROSS_SQUARE_FEET')] = data['GROSS_SQUARE_FEET'].fillna(0)
```

```python
data[('LAND_SQUARE_FEET')] = data['LAND_SQUARE_FEET'].fillna(0)
```

```python
data = data[(data['SALE_PRICE'] > 100000) & (data['SALE_PRICE'] < 5500000)]
```

```python
data = data[(data['TOTAL_UNITS'] > 0) & (data['TOTAL_UNITS'] < 60)] 
```

```python
data = data[data['GROSS_SQUARE_FEET']>0]
```

```python
data = data[data['LAND_SQUARE_FEET']>0]
```

```python
data = data[data['TOTAL_UNITS']>0]
```

```python
data['SALE_PRICE'].describe
```

```python
data['GROSS_SQUARE_FEET'].describe
```

```python
data['LAND_SQUARE_FEET'].describe
```

```python
data.columns
```

```python
features = ['BOROUGH', 'BLOCK','LOT', 'ZIP_CODE', 'TOTAL_UNITS','LAND_SQUARE_FEET']
```

```python
X = data[features]
```

```python
y = data['SALE_PRICE']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 60)
```

```python
model=GradientBoostingRegressor()
```

```python
model.fit(X_train, y_train)
```

```python
#coefficient = model.coef_
```

```python
#intercept = model.intercept_
```

```python
model.score(X_test, y_test)
```

```python
y_predicted = model.predict(X_test)
```
